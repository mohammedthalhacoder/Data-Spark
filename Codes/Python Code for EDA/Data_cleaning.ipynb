```python
import pandas as pd
from datetime import datetime

# import packages as needed
```

```python
df_cus = pd.read_csv(r"F:\GUVI\Project\Data Spark\Data\Customers.csv", encoding="ISO-8859-1")

# read the csv file by file path and store to the variable
```

```python
df_cus
```

```python
df_cus.drop_duplicates()
df_cus

# remove the duplicates and print the data
```

```python
df_cus.info()

# chech the information in the data and find the null
```

```python
df_cus[df_cus['State Code'].isnull()]

# finded the null values in particular columns
```

```python
df_cus['State Code'] = df_cus['State Code'].fillna('NAP')

# directly replace on the column because only the NaN found for same state
```

```python
df_cus.info()

# again checking the null values and informations
```

```python
df_cus['Birthday'] = pd.to_datetime(df_cus['Birthday'], format = "%m/%d/%Y")

# changing the dtypes to datetime to calculate the age
```

```python
df_cus
```

```python
today = datetime.today()
today

# finding today date and time to calculate the age
```

```python
df_cus['Age'] = today.year - df_cus['Birthday'].dt.year

# minus only the year from today and birthday column and store it to new columns
```

```python
df_cus
```

```python
df_cus.info()
```

```python
df_cus['City'] = df_cus['City'].str.title()
```

```python
df_cus
```

```python
df_cus.to_csv("Cleaned_Customers_data.csv")

# coverted to csv
```
